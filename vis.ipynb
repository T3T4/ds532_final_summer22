```python
df2=pd.read_sql('SELECT Comments_avg from Comments',con=mysql)
df3=pd.read_sql('SELECT Likes_avg from Likes',con=mysql)
df4=pd.read_sql('SELECT Shares_avg from Shares',con=mysql)
df5=pd.read_sql('SELECT Subscribers_count from Subscribers',con=mysql)
df6=pd.read_sql('SELECT Views_avg from Views',con=mysql)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(15,8))
sns.histplot(data=df2,x="Comments_avg", bins=50,color="#bfc934")
plt.savefig('comments_7-16.png')
```

```python
df3=pd.read_sql('SELECT Likes_avg from Likes',con=mysql)
plt.figure(figsize=(15,8))
sns.histplot(data=df3,x="Likes_avg",bins=50, color="#D533A8",log_scale=True)
plt.savefig('likes_7-16.png')
```

```python
df4=pd.read_sql('SELECT Shares_avg from Shares',con=mysql)
plt.figure(figsize=(15,8))
sns.histplot(data=df4, x="Shares_avg",bins=50,color="#329410", log_scale=True)
plt.savefig('shares_7-16.png')
```

```python
df5=pd.read_sql('SELECT Subscribers_count from Subscribers',con=mysql)
plt.figure(figsize=(15,8))
sns.histplot(data=df5, x="Subscribers_count",color="#229B89")
plt.savefig('subscribers_7-16.png')
```

```python
df6=pd.read_sql('SELECT Views_avg from Views',con=mysql)
plt.figure(figsize=(15,8))
sns.histplot(data=df6, x="Views_avg",bins=50, color="gray")
plt.savefig('views_7-16.png')
```
